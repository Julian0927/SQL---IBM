# ***Task 3: Create the database connection.***

`Ibm_db` API uses the IBM Data Server Driver for ODBC and CLI APIs to connect to IBM DB2 and Informix.


```python
#Create database connection
#DO NOT MODIFY THIS CELL. Just RUN it with Shift + Enter
dsn = (
    "DRIVER={0};"
    "DATABASE={1};"
    "HOSTNAME={2};"
    "PORT={3};"
    "PROTOCOL={4};"
    "UID={5};"
    "PWD={6};"
    "SECURITY={7};").format(dsn_driver, dsn_database, dsn_hostname, dsn_port, dsn_protocol, dsn_uid, dsn_pwd,dsn_security)
```

```python
try:
    conn = ibm_db.connect(dsn, "", "")
    print ("Connected to database: ", dsn_database, "as user: ", dsn_uid, "on host: ", dsn_hostname)

except:
    print ("Unable to connect: ", ibm_db.conn_errormsg() )
```

## ***Task 6: Query data in the table.***

In this step we will retrieve data we inserted into the INSTRUCTOR table.


```python
# Construimos la consulta que recupera todos los renglones de la tabla INSTRUCTOR en formato SQL.
select_Query = "select * from INSTRUCTOR"

# Ejecutamos la sentencia anterior para recuperar realmente 
select_Stmt = ibm_db.exec_immediate(conn, select_Query)
select_Stmt
```

```python
dropStmt = ibm_db.exec_immediate(conn, "drop table INSTRUCTOR")


createQuery = "create table INSTRUCTOR(id INTEGER PRIMARY KEY NOT NULL, fname VARCHAR(20), lname VARCHAR(20), city VARCHAR(20), ccode CHAR(2))"
createStmt = ibm_db.exec_immediate(conn, createQuery)


insertQuery_1 = "insert into INSTRUCTOR values (1, 'Rav', 'Ahuja', 'TORONTO', 'CA')"
insertStmt = ibm_db.exec_immediate(conn, insertQuery_1)


insertQuery_2 = "insert into INSTRUCTOR values (2, 'Raul', 'Chong', 'Markham', 'CA')"
insertStmt = ibm_db.exec_immediate(conn, insertQuery_2) 


insertQuery_34 = "insert into INSTRUCTOR values (3, 'Andres', 'Chong', 'Markham', 'CA'), (4, 'Julian', 'Uribe', 'Medellin', 'CO')"
insertStmt = ibm_db.exec_immediate(conn, insertQuery_34)


select_Query = "select * from INSTRUCTOR"
select_Stmt = ibm_db.exec_immediate(conn, select_Query)
```

```python
#Fetch the rest of the rows and print the ID and FNAME for those rows
while ibm_db.fetch_row(select_Stmt) != False:
   print("ID:",  ibm_db.result(select_Stmt, 0), "FNAME:",  ibm_db.result(select_Stmt, "FNAME"))
```

## ***Task 7: Retrieve data into Pandas***

In this step we will retrieve the contents of the INSTRUCTOR table into a Pandas dataframe.


```python
import pandas
import ibm_db_dbi
```

```python
# connection for pandas
# conn = ibm_db.connect(dsn, "", "")
p_conn = ibm_db_dbi.Connection(conn) 
```

<img src="https://ibm.box.com/shared/static/j5yjassxefrjknivfpekj7698dqe4d8i.jpg" align="center">

```python
#query statement to retrieve all rows in INSTRUCTOR table
select_Query = "select * from INSTRUCTOR"

#retrieve the query results into a pandas dataframe
pandas_df = pandas.read_sql(select_Query, p_conn)
pandas_df 
```

```python
#print just the LNAME for first row in the pandas data frame
pandas_df.LNAME[0]
```
